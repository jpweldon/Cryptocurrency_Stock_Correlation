### Import the required libraries and dependencies

```python
#import libaries 
import numpy as np
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

### Import SPY data from "Resources" folder
1. closing prices
2. daily returns

```python
# Using the Pandas read_csv function and the Path module, read in the SPY closing prices
# from the "Market_Data" notebook and create a DataFrame
spy_closing_prices = pd.read_csv(
    Path('../Resources/spy_closing_prices.csv')
)

# Assign descriptive headers to the columns
columns = ['Date', 'SPY_Close']
spy_closing_prices.columns = columns

# Set the "Date" column as the index
spy_closing_prices = spy_closing_prices.set_index('Date')

# Confirm the changes by showing the first and last 5 rows of the DataFrame
display(spy_closing_prices.head())
display(spy_closing_prices.tail())
```

```python
# Using the Pandas read_csv function and the Path module, read in the SPY daily returns
# from the "Market_Data" notebook and create a DataFrame
spy_daily_returns = pd.read_csv(
    Path('../Resources/spy_daily_returns.csv')
)

# Assign descriptive headers to the columns
columns = ['Date', 'SPY_Daily_Returns']
spy_daily_returns.columns = columns

# Set the "Date" column as the index
spy_daily_returns = spy_daily_returns.set_index('Date')

# Confirm the changes by showing the first and last 5 rows of the DataFrame
display(spy_daily_returns.head())
display(spy_daily_returns.tail())
```

### Import VIX data from "Resources" folder
1. closing prices
2. daily change
3. cumulative daily changes

```python
# Using the Pandas read_csv function and the Path module, read in the VIX closing prices
# from the "Volatility_Data" notebook and create a DataFrame
vix_closing_prices = pd.read_csv(
    Path('../Resources/vix_closing_prices.csv')
)

# Assign descriptive headers to the columns
columns = ['Date', 'VIX_Close']
vix_closing_prices.columns = columns

# Set the "Date" column as the index
vix_closing_prices = vix_closing_prices.set_index('Date')

# Confirm the changes by showing the first and last 5 rows of the DataFrame
display(vix_closing_prices.head())
display(vix_closing_prices.tail())
```

```python
# Calculate daily change of VIX closing prices
vix_daily_change = vix_closing_prices.pct_change().dropna()

# Change column header name
columns = ['VIX_Change']
vix_daily_change.columns = columns

# Check first 5 rows
vix_daily_change.head()
```

```python
# Create new DataFrame of cumulative sum of VIX daily changes
vix_cumulative_return = (1 + vix_daily_change).cumprod()
columns = ['VIX_cum_ret']
vix_cumulative_return.columns = columns

# Verify function worked by reviewing last few lines of DataFrame
vix_cumulative_return.tail()
```

### Import crypto price data from "Resources" folder
1. closing prices
2. daily returns

```python
# Using the Pandas read_csv function and the Path module, read in the crypto closing prices
# from the "Crypto_Data" notebook and create a DataFrame
crypto_historic_closing = pd.read_csv (
    Path('../Resources/crypto_historic_closing.csv')
)

# Assign descriptive headers to the columns
columns = ['Date', 'BTC_Close', 'ETH_Close', 'LTC_Close']
crypto_historic_closing.columns = columns

# Set the "Date" column as the index
crypto_historic_closing = crypto_historic_closing.set_index('Date')

# Confirm the changes by showing the first and last 5 rows of the DataFrame
display(crypto_historic_closing.head())
display(crypto_historic_closing.tail())
```

```python
# Using the Pandas read_csv function and the Path module, read in the crypto daily returns
# from the "Crypto_Data" notebook and create a DataFrame
crypto_historic_daily_returns = pd.read_csv(
    Path('../Resources/crypto_historic_daily_returns.csv')
)

# Assign descriptive headers to the columns
columns = ['Date', 'BTC_Daily_Returns', 'ETH_Daily_Returns', 'LTC_Daily_Returns']
crypto_historic_daily_returns.columns = columns

# Set the "Date" column as the index
crypto_historic_daily_returns = crypto_historic_daily_returns.set_index('Date')

# Confirm the changes by showing the first and last 5 rows of the DataFrame
display(crypto_historic_daily_returns.head())
display(crypto_historic_daily_returns.tail())
```

# Visualizing imported data with plots

```python
# Create the plot for the daily returns spy
plot_daily_returns_spy = spy_daily_returns.hvplot.box(
    title ='Total Daily Returns',
    height=400,
    width=800,
    xlabel='SPY'
)

# Visualize the plot
plot_daily_returns_spy
```

```python
# Create box plots for daily returns for the three cryptocurrencies
plot_daily_returns_crypto = crypto_historic_daily_returns.hvplot.box(
    height=400,
    width=800,
    xlabel='Crypto Currency',
    ylabel='Daily Return',
    title='Crypto Daily Return of BTC, ETH, LTC from 2019-7-1 - 2021-7-1'
)

# Visualize the plot
plot_daily_returns_crypto
```

```python
# Create an overlay box plot to include the SPY daily returns
overlay_plot = plot_daily_returns_crypto * plot_daily_returns_spy

# Display the box plots
overlay_plot
```

```python
# Create new DataFrame of cumulative returns
spy_cumulative_return = (1 + spy_daily_returns).cumprod()
columns = ['SPY_cum_ret']
spy_cumulative_return.columns = columns

# Verify function worked by reviewing last few lines of DataFrame
spy_cumulative_return.tail()
```

```python
# Plot the SPY Daily Returns
spy_cumulative_return.plot(
    figsize = (15,10),
    title = 'SPY - Cumulative Returns',
    ylabel = 'Cumulative Return'
)
```

```python
# Overlay the cumulative returns for the SPY and the VIX daily changes
# Use axes objects (ax) to overlay two DataFrames
ax = spy_cumulative_return.plot(
    figsize = (15,7),
    title = 'SPY vs VIX',
    ylabel='Cumulative Return'
)

# Call the second DataFrame to the plot
vix_cumulative_return.plot(ax=ax)
```

```python
# Calculate cumulative returns for all three cryptos over period
crypto_cumulative_returns = (1 + crypto_historic_daily_returns).cumprod()

# Assign proper headers for each column in DataFrame
columns = ['BTC_cum_ret','ETH_cum_ret','LTC_cum_ret']
crypto_cumulative_returns.columns = columns

# Display last 5 rows to verify function worked and then plot the cumulative returns
display(crypto_cumulative_returns.tail())
crypto_cumulative_returns.plot(
    figsize = (15,10), 
    title = 'Crypto Cumulative Returns'
)
```

```python
# Use pandas concat() function to create new DataFrame to include all assets organized by date
overlay_dates = pd.concat(
    [
        crypto_cumulative_returns, 
        spy_cumulative_return, 
        vix_cumulative_return
    ], 
    axis = 1, 
    join = 'inner'
)

# Plot the new DataFrame to visualize all cumulative returns and VIX data
overlay_dates.plot(
    figsize = (15,7),
    title = 'Cumulative Returns - All Assets'
)
```

```python
# Separate plot into two sections to better visualize correlations between assets
# Create DataFrame for first period using loc[] fucntion
first_part_overlay = overlay_dates.loc['2019-07-01':'2020-12-01']

# Create DataFrame for second period using loc[] fucntion
second_part_overlay = overlay_dates.loc['2020-12-02':'2021-07-01']

# Display last 5 rows of each DataFrame using the tail() function to confirm split worked
display(first_part_overlay.tail())
display(second_part_overlay.tail())
```

```python
# Plot the cumulative return data from the first period
first_part_overlay.plot(
    figsize = (15,10),
    title = 'Cumulative Returns - July 1st 2019 to December 1st 2020',
    ylabel = 'Return'
)
```

```python
# Plot the cumulative return data from the second period
second_part_overlay.plot(
    figsize = (15,10),
    title = 'Cumulative Returns - December 2nd 2020 to July 1st 2021',
    ylabel = 'Return'
)
```

```python
# Assign column headers for the correlation datatable
columns = ['BTC','ETH','LTC','SPY','VIX']
overlay_dates.columns = columns

# Generate correlation statistics for all assets using the corr() function 
overlay_dates.corr()
```
